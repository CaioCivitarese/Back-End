# 1) Variáveis
Uma **variável** funciona como uma **caixa com nome**, e você pode colocar dentro dela diferentes tipos de dados, como números, palavras ou até mesmo listas de dados. Essa **caixa com nome** ocupa um espaço na memória do computador.

Python é case-sensitive: então “A” e “a” são símbolos diferentes e se referem a diferentes variáveis.

Vamos dizer que você desejar armazenar na memória do computador sua idade. No mundo da programação, você criaria uma variável chamada, por exemplo, **idade** e diria que elá contém o número **21** (ou qualquer que seja sua idade):

```python
# Atribui o valor 21 para a variável idade.

idade = 21
```

Agora, toda vez que você desejar saber ou mudar sua idade no programa, você deve utilizar a variável **idade**. Para verificar o valor armazenado em uma variável basta utilizar o comando "print". Exemplo:

```python
# Aribui o valor 21 para a variável idade.

idade = 21

# Imprime na tela o valor da variável idade.

print(idade)
```

Existem diferentes tipo de váriaveis.

Imagine que está contando a quantidade de moedas de real que estão em sua carteira. Os números que você conta são sempre inteiros, certo? Tipo 1, 2, 3... Então, quando você usa uma variável do tipo inteiro na programação, está guardando números inteiros, sem casas decimais. Exemplo:

```python
# Atribui o valor 10 para a variável quantidade.

quantidade = 10

# Imprime na tela o valor da variável.

print(quantidade)

# Imprime o tipo da variável

type(quantidade)
```

Agora, pense no valor financeiro representado pelas moedas, ele pode ser um número não inteiro, como R$ 3,50. Esse tipo de número, com casas decimais, é chamado de ponto flutuante. Na programação, você usaria uma variável do tipo float para armazenar esses valores. Exemplo:

```python
# Atribui o número 1.75 (ponto e não vírgula) à variável valor.

valor = 1.75  # valor é uma variável do tipo float.
print(valor)  # Imprime na tela o conteúdo da variável valor.
type(valor)   # Imprime na tela o tipo da variável valor.
```

Agora, pense nas palavras e frases. Quando você desejar guardar texto, como o nome de uma pessoa, deve utilizar uma variável do tipo str (string). Strings são usadas para representar sequências de caracteres.

```python
# Atribui a string 'Carlos Santos' à variável nome.

nome = "Carlos Santos" # nome é uma variável do tipo string.
print(nome)     # Imprime na tela o conteúdo da variável nome.
type(nome)      # Imprime na tela o tipo da variável nome.
```

E se tivermos um conjunto de nomes ou um conjunto de valores? Para isso temos a lista. Pense na lista como uma coleção de dados, como uma lista de compras. Em programação, uma lista é uma estrutura de dados que permite armazenar diversos valores sob uma mesma variável.

```python
nomes = ["Carlos", "Simone", "Marcelo"] # nomes é uma variável do tipo list.
print(nomes)                            # Imprime o valor da variável.
print(nomes[0])                         # O primeiro elemento da lista está na posição 0, o segundo na posição 1, e assim por diante.
print(nomes[1])
type(nomes)                             # Imprime o tipo da variável.
```

# 2) Operadores aritméticos no python

Em Python, existem vários tipos de operadores que realizam diferentes operações em valores ou variáveis.

```python
# Operadores aritméticos com as variáveis valor1 e valor2.

valor1 = 9
valor2 = 2

soma = valor1 + valor2              # Adição
subtracao = valor1 - valor2         # Subtração
multiplicacao = valor1 * valor2     # Multiplicação
divisao = valor1 / valor2           # Divisão
divisao_inteira = valor1 // valor2  # Divisão Inteira
resto = valor1 % valor2             # Resto da divisão inteira
potencia = valor1 ** valor2         # Potência

# Exibindo os resultados.
print("Soma:", soma)
print("Subtração:", subtracao)
print("Multiplicação:", multiplicacao)
print("Divisão:", divisao)
print("Divisão Inteira:", divisao_inteira)
print("Resto da Divisão:", resto)
print("Potência:", potencia)

# Modificando o valor de uma variável.

valor1 = 10 # Uma vez criada a variável, posso modificar o seu valor.
soma = valor1 + valor2  # Cálculando o novo valor para a variável soma.

print("Novo valor da variável valor1:", valor1)
print("Novo valor da variável soma:", soma)

```

```python
# O uso de parênteses segue a mesma regra que aprendemos na matemática.
# O uso de parênteses facilita o ententimento das fórmulas.

peso = 80
altura = 1.60
imc = (peso / (altura * altura))  # imc é o índice de massa corporal de uma pessoal.

print("IMC:", imc)
```

# 3) Operadores lógicos em Python

Os operadores lógicos são usados para combinar ou modificar expressões booleanas em programação. Em Python, existem três operadores lógicos principais: and, or e not. Vamos entender cada um deles:

O operador **and** retorna **True** se ambas as expressões booleanas forem verdadeiras; caso contrário, retorna **False**. Exemplo:

```python
x = True
y = False

resultado = x and y
print(resultado)
```

Neste exemplo, resultado será **False** porque pelo menos uma das variáveis, no caso **y**, é falsa.

O operador **or** retorna **True** se pelo menos uma das expressões booleanas for verdadeira; retorna **False** apenas se ambas forem falsas.

```python
a = True
b = False

resultado = a or b
print(resultado)
```

Aqui, resultado é **True** porque pelo menos uma das variáveis, no caso **a**, é verdadeira.

O operador **not** é usado para inverter o valor de uma expressão booleana. Se a expressão for **True**, **not** a torna **False**; se a expressão for **False**, **not** a torna **True**.

```python
a = True
b = False

c = a or b  # c será True.

resultado = not c
print(resultado)
```

Neste caso, resultado é **False** porque **not c** inverte o valor de **c**.

# 4) Estruturas de controle - Condicionais

As estruturas de controle condicionais são usadas em programação para tomar decisões com base em condições específicas. Em Python, as estruturas de controle condicionais são implementadas principalmente através dos blocos **if / else** e, opcionalmente, **elif** (abreviação de "else if").

```python
idade = 18

if (idade >= 18): # Se idade for maior ou a igual a 18.
    print("Você é maior de idade.")
else: # Se a condição acima não for vedadeira.
    print("Você é menor de idade.")
```

Exemplo com operadores lógicos e operadores aritméticos:

```python
idade = 22
temPermissao = True

if ((idade >= 18) and temPermissao): # Retorna verdadeiro se as duas condições forem verdadeiras.
    print("Pode entrar.")
else:  # Se pelo menos uma das expressões for falsa.
    print("Não pode entrar.")
```

Exemplo utilizando o **elif**:

```python
# Exemplo usando elif para classificar o clima.

clima = "ensolarado"

if (clima == "chuvoso"): # Se clima for igual (== e não =) a "chuvoso".
    print("Lembre-se de levar um guarda-chuva!")
elif (clima == "ensolarado"): # Se clima for igual a "ensoladorado".
    print("Ótimo dia para atividades ao ar livre.")
elif (clima == "nublado"): # Se clima for igual a "nublado".
    print("Talvez não seja tão ensolarado, mas ainda bom.")
else: # Se nenhuma das condições acima for verdadeira.
    print("Condição meteorológica desconhecida.")
```

# 5) Estrutura de controle - Loop

As estruturas de controle de loops são utilizadas para repetir um bloco de código várias vezes. Em Python, há duas formas principais de implementar loops: comandos **for** e **while**.

O loop **for** é usado quando se conhece o número exato de iterações. Ele itera sobre uma sequência (como uma lista, string, ou intervalo numérico) e executa o bloco de código para cada elemento na sequência. Em python, uma string é considerada uma lista de caracteres.

No exemplo abaixo, o loop **for** itera sobre cada elemento da lista frutas e imprime cada uma das frutas.

```python
# Exemplo de loop for para iterar sobre uma lista.

frutas = ["maçã", "banana", "uva"]

for fruta in frutas:
    print(fruta)
```

```python
# Exemplo de loop for para iterar sobre uma sequência de números inteiros.

for i in range(10):    # Imprime 10 números começando em 0 e com passo 1. 
    print(i)
```

```python
# Exemplo de loop for para iterar sobre uma sequência de números inteiros.

for i in range(2, 6):    # Imprimee os números de 2 a 5, exceto o 3. O comando range é aberto à direita, logo, começa em 0 e vai até 5 (valor inteiro anterior a 6). 
    if (i != 3):
        print(i)
    else:
        print('É 3')
```

O bloco de código definido pelo comando **while** é executado repetidamente enquanto a condição for verdadeira.

```python
# Exemplo de loop while para contar de 0 a 5.

contador = 0

while (contador < 5):
    print(contador)
    contador = contador + 1
```

Neste exemplo, o loop **while** imprime o valor da variável **contador** enquanto ela for menor que 5.

# 6) Funções em Python

Em programação, uma **função** é um **bloco de código reutilizável que executa uma tarefa específica**.
As funções ajudam a organizar o código, torná-lo mais legível e facilitar a manutenção.
Em Python, funções são definidas usando a palavra-chave **def**.
Vamos explorar os principais elementos de uma função:

Sintaxe básica de uma função:

```python
def nomeDaFuncao (parametro1, parametro2): # Uma função pode apresentar nenhum, um ou mais valores de entrada.
    # Corpo da função.
    # Realiza operações com os parâmetros e/ou variáveis locais.
    return (resultado) # resultado (poderia ser outro nome) é uma varável que armazena o valor que será retornado pela função.
```

**nomeDaFuncao**: Nome escolhido para a função.

**parametro1**, **parametro2**: Parâmetros que a função recebe.

**return (resultado)**: A palavra-chave **return** indica o valor que a função retorna. Uma função pode ou não ter uma declaração de retorno.

Exemplo de função sem valor de retorno:

```python
# Esta função imprime uma saudação.

def apresentarSaudacao(nome, sobrenome):
    print(f"Olá, {nome}, {sobrenome}!")

# Chamando a função.
apresentarSaudacao("Carlos", "Santos")
```

Exemplo com dois parâmetros e com valor de retorno:

```python
# Esta função retorna a soma de dois números.

def calcularSoma(x, y):
    resultado = x + y
    return (resultado)

# Declara as variáveis valor1 e valor2
valor1 = 3
valor2 = 5

# Declara a variável resultadoSoma.
resultadoSoma = 0

# Chama a função e armazena o resultado retornado por ela na variável resultadoSoma.
resultadoSoma = calcularSoma(valor1, valor2)

# Imprime o valor da variável resultadoSoma.
print(resultadoSoma)
```

## 7.2 Carregando o Conjunto de Dados

O comando **read_csv** do **Pandas** permite disponibilizar em uma estrutura chamada **data frame** (semelhante a uma tabela) os dados armazenados em um arquivo no padrão **CSV** (Comma-Separated Values).

```python
# O arquivo se chama DadosRJ.csv e está no subdiretório Dados.

caminhoArquivo = 'Arquivos/DadosRJ.csv'

# O arquivo DadosRJ.csv contém os dados de código, população e PIB per capta dos municípios do Estado do Rio de Janeiro.
# CSV (Comma Separated Values) é um formato de arquivo de texto simples usado para armazenar dados tabulares,
# onde cada linha representa um registro e os valores são separados por vírgulas (ou outros delimitadores).
# 'pd.read_csv()' é um métodos usado para ler arquivos CSV e carregar seus dados em uma estrutura chamada data frame.

# A variável dadosRJ é um data frame e irá armazenar os dados que estão no arquivo 'DadosRJ.csv'.
dadosRJ = pd.read_csv(caminhoArquivo, sep = ";", decimal = ",") # Separador de coluna (parâmetro sep) e de casas decimais (parâmetro decimal) do arquivo.
```

# 8) Biblioteca Matplotlib

O **Matplotlib** é uma biblioteca de visualização de dados para Python.  
Essencialmente, ela permite criar gráficos e visualizações de dados em 2D, com uma grande variedade de tipos de gráficos disponíveis, como gráficos de linha, barras, pizza, histogramas e gráficos de dispersão.  
O **Seaborn** é uma biblioteca de visualização de dados em Python, construída sobre o Matplotlib, que facilita a criação de gráficos estatísticos atraentes e informativos. 
Os links abaixo apresentam tutorais para as duas bibliotecas:  
https://matplotlib.org/stable/users/explain/quick_start.html  
https://seaborn.pydata.org/tutorial/introduction.html


```python
# Exemplo de gráfico de barras contendo os 10 municípios com maior população do esstado do Rio de Janeiro.

# Importa o módulo pyplot da biblioteca matplotlib.
import matplotlib.pyplot as plt

# Seleciona os municípios com as 10 maiores populações.
# Primeoro ordena do maior para o menor e a seguir seleciona as 10 maiores.
top10 = dadosRJ.sort_values(by='populacao', ascending = False).head(10)

# Definição do apresentação do gráfico.

# Cria uma área para o gráfico e define seu tamanho em polegadas, 10 de largura e 6 de altura.
plt.figure(figsize=(10, 6))

# A função barh permite criar gráficos de barras horizontais e a função bar gráficos de barras verticais.
# O primeiro parâmetro é a lista de rótulos e o segundo parâmetro a lista de valores.
# .astype(str) é utilizado para que a soluna 'codigo' seja tratado como sendo do tipo string.
# color defina tanto a cor da borda quanto a cor de fundo das barras.
plt.barh(top10['codigo'].astype(str), top10['populacao'], color = 'steelblue')

# Define o título do eixo horizontal.
plt.xlabel('População')

# Define o título do gráfico
plt.title('10 Municípios mais populosos do Estado do Rio de Janeiro')

# Define que o município de maior população ficará no topo.
plt.gca().invert_yaxis()

# Adiciona uma grade ao gráfico.
plt.grid(True)

# Apresenta o gráfico.
plt.show()
```

```python
# Exemplo de gráfico de setores contendo os 10 municípios com maior população do esstado do Rio de Janeiro.

# Importa o módulo pyplot da biblioteca matplotlib.
import matplotlib.pyplot as plt

# Seleciona os municípios com as 10 maiores populações.
# Primeoro ordena do maior para o menor e a seguir seleciona as 10 maiores.
top10 = dadosRJ.sort_values(by = 'populacao', ascending = False).head(10)

# Definição do apresentação do gráfico.
plt.figure(figsize = (8, 8))  # Tamanho do gráfico.

# A função pie permite criar um gráfico de setores.
plt.pie(
    top10['populacao'],                # Valores.
    labels = top10['codigo'],          # Rótulos.
    autopct = '%1.1f%%',               # Mostra porcentagem com 1 casa decimal.
    startangle = 80,                   # Ângulo inicial de rotação do gráfico.
    colors = plt.cm.Paired.colors      # Paleta de cores (opcional).
)

# Define o título do gráfico.
plt.title('Distribuição Populacional dos 5 Maiores Municípios do RJ')

# Apresenta o gráfico
plt.show()
```

Nesta seção, os alunos têm a chance de aplicar o que aprenderam utilizando o arquivo filmes.csv.

```python
# O arquivo filmes.csv (baseado no data set MovieLens, conjunto de dados amplamente utilizado em pesquisas e aplicações de sistemas de recomendação)
# contém avaliações feitas por usuários a milhares de filmes, onde cada linha representa uma nota atribuída por um usuário a um filme específico.
# Usuário - Identificador anonimizado do usuário.
# Filme - Identificador anonimizado do filme.
# Nota - 1 a 5.
# Timestamp - Data/hora da avaliação (epoch).
    
# Separador de coluna do arquivo é tab (\t).
# As notas são números inteiros e por isso não definimos o parâmetro decimal no read_csv.
# O arquivo não contém nome das colunas e por isso precisamos definir o parâmetro names.

import pandas as pd

caminhoArquivo = 'Arquivos/filmes.csv'
atributos = ['Usuario', 'Filme', 'Nota', 'Timestamp']
dadosFilme = pd.read_csv(caminhoArquivo, sep = "\t", names = atributos)
print(dadosFilme.head(10))
```

```python
# Calcule: 1) o total de registros; 2) o número de usuários que o arquivo possui; 3) o número de filmes avaliados.

totalRegistros = dadosFilme.shape[0]
totalUsuarios = len(dadosFilme['Usuario'].unique()) # 'unique()' retorna uma lista sem repetição dos usuários e 'len' o número de elementos desta lista.
totalFilmes = len(dadosFilme['Filme'].unique())     # 'unique()' retorna uma lista sem repetição dos filmes e 'len' o número de elementos desta lista.

print(f'Total de registros {totalRegistros}.')
print(f'Total de usuários {totalUsuarios}.')
print(f'Total de filmes {totalFilmes}.')
```

```python
# Calcule o número de notas (frequência) dadas por nota.

import numpy as np

totalPorNota = dadosFilme['Nota'].value_counts()   # Para cada valor único presente na coluna ou Nota, a função value_counts() conta quantas vezes ele aparece.
print(totalPorNota)                                # tatal de avaliações realizadas para cada possível valor de nota.
type(totalPorNota)                                 # totalPorNota é um tipo de dados do Python chamado Series.
```

```python
# Crie um data frame novo contendo somente as colunas 'Filme' e 'Nota'.

datosFilmeSelecionado =  dadosFilme[['Filme', 'Nota']]
datosFilmeSelecionado.head()
```

```python
# Calcule: 1) a médias das notas fornecidas; 2) total de avaliações acima da média; 3) a média das notas por filme;
# 4) o total de avalições e a média das notas por filme.

mediaNotas = dadosFilme['Nota'].mean()

totalAcimaMedia = len(dadosFilme[(dadosFilme.Nota > mediaNotas)])

# Agrupa por 'Filme' e depois calcula para cada filme a média das notas (usa a coluna 'Nota').
mediaNotaPorFilme = dadosFilme.groupby(['Filme'])['Nota'].mean() # mediaNotaPorFilme é um tipo de dados do Python chamado Series.

# Seleciona as colunas 'Filme' e 'Nota' e a seguir agrupa por 'Filme'.
# O método agg permite aplicar funções ao agrupamento criado.
# No exemplo, as funções 'count' (contagem) e 'mean' (média) são aplicadas à coluna 'Nota' dentro do agrupamento criado (por 'Filme').
consolidadoFilme = dadosFilme[['Filme', 'Nota']].groupby(['Filme']).agg(['count', 'mean'])

print(f"A média das notas é {mediaNotas:.2f}.")
print(f"A quantidade de avaliações acima da média é {totalAcimaMedia}.")

print("")
print("#### Média das notas por filme. ###")
print(mediaNotaPorFilme)

print("")
print("### Total de notas e média das notas por filme. ###")
print(consolidadoFilme)
```

```python
# Apresente um gráfico de barras das notas.

import matplotlib.pyplot as plt
import seaborn as sns

counts = (dadosFilme['Nota'].value_counts(normalize = True)*100).reset_index()
counts.sort_values(['Nota'], ascending = True)

sns.set_theme(style = 'darkgrid')
sns.barplot(data = counts, x = 'Nota', y = 'proportion')
plt.title('Distribuição das notas')
plt.xlabel('Nota')
plt.ylabel('%')
plt.show()
```

```python
# Apresente um gráfico de setores das notas.

import matplotlib.pyplot as plt
import seaborn as sns

counts = dadosFilme['Nota'].value_counts().values
notas = dadosFilme['Nota'].value_counts().index

sns.set_theme(style = 'darkgrid')
plt.pie(x = counts, labels = list(notas), autopct='%1.1f%%')
plt.title('Distribuição das notas')
plt.axis('equal')
plt.show()
```
